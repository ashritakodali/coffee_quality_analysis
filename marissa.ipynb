```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Read in arabica date
arabica = pd.read_csv("df_1_arabica.csv")
arabica
```

```python
# Read in robusta data
robusta = pd.read_csv("df_1_robusta.csv")
robusta
```

```python
# Set bean type in both tables for distinction between observations
arabica['bean_type'] = 'arabica'
robusta['bean_type'] = 'robusta'
```

```python
# Concatenate to join tables
coffee = pd.concat([arabica, robusta], ignore_index=True)
```

```python
# Make all variable names lowercase and have underscores if there are spaces
coffee.columns = coffee.columns.str.lower().str.replace(' ', '_')

# Fix columns that did not transform properly
coffee['bitter_and_sweet'] = coffee['bitter_/_sweet']
coffee['salt_and_acid'] = coffee['salt_/_acid']
coffee['fragrance_and_aroma'] = coffee['fragrance_/_aroma']
```

```python
# Remove unneccessary columns
columns_to_drop = ['na', 'fragrance_/_aroma', 'bitter_/_sweet', 'salt_/_acid']
coffee = coffee.drop(columns_to_drop, axis=1)
coffee
```

```python
# Get info about the dataframe to assist with cleaning
coffee.info()
```

### Convert variables with numbers to float dtype (if not already done)

```python
coffee['bag_weight'] = coffee['bag_weight'].str.replace(' kg','').astype(float)
```

```python
coffee['moisture'] = coffee['moisture'].str.replace('%', '').astype(float) / 100
```

```python
coffee['category_one_defects'] = coffee['category_one_defects'].str.replace(' full defects','').astype(float)
```

```python
coffee['category_two_defects'] = coffee['category_two_defects'].str.replace(' full defects','').astype(float)
```

### Convert dates to date dtype

```python
from dateutil import parser

coffee['grading_date'] = coffee['grading_date'].apply(parser.parse)
```

```python
coffee['expiration'] = coffee['expiration'].apply(parser.parse)
```
